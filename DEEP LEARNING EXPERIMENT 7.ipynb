```python
import tensorflow as tf 
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from keras.layers import Conv2D, Dense
from keras.models import Sequential

```

```python
image = plt.imread('my image 2 for resume .jpeg')
model = keras.Sequential()
identity_filter = np.array([[0, 0, 0],
                             [0, 1, 0],
                             [0, 0, 0]], dtype=np.float32)
```

```python
sobel_filter = np.array([[-1, -2, -1],
                          [0, 0, 0],
                          [1, 2, 1]], dtype=np.float32)

gaussian_filter = np.array([[1, 2, 1],
                             [2, 4, 2],
                             [1, 2, 1]], dtype=np.float32) / 16.0  # Normalize the filter

#image=np.expand_dims(image,axis=0)
```

```python
model.add(keras.layers.Conv2D(1, (3, 3), input_shape=image.shape[1:], padding='valid', use_bias=False,
                              kernel_initializer=keras.initializers.Constant(value=identity_filter)))
identity_feature_map = model.predict(image)

model.add(keras.layers.Conv2D(1, (3, 3), input_shape=image.shape[1:], padding='valid', use_bias=False,
                              kernel_initializer=keras.initializers.Constant(value=sobel_filter)))
sobel_feature_map = model.predict(image)
```

```python
model = keras.Sequential()
model.add(keras.layers.Conv2D(1, (3, 3), input_shape=image.shape[1:], padding='valid', use_bias=False,
                              kernel_initializer=keras.initializers.Constant(value=gaussian_filter)))
gaussian_feature_map = model.predict(image)

```
